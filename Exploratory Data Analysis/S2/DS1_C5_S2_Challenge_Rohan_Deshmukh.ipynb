```python
import matplotlib.pyplot as plt             # to visualize
from tabulate import tabulate               # to print the table
import matplotlib as mat                    # to visualize 
import seaborn as sns                       # to visualize
import pandas as pd                         # for data reading
import numpy as np                          # for numerical computation
```

```python
df = pd.read_excel('DS1_C5_S2_Income_Data_Challenge.xlsx')
df
```

```python
l=list(df.columns)
for i in range(len(l)):
    print(l[i])
    print(df[l[i]].unique())
    print()
```

```python
df.info()
```

```python
def col(df):
    x=df.select_dtypes(exclude='object')
    y=df.select_dtypes(include='object')
    x1=list(x.columns)
    y1=list(y.columns)
    return x1,y1


numerical,categorical=col(df)
print("Numerical")
print(numerical)
print("Categorical")
print(categorical)
```

```python
# Task 1
def info_cat(c):
    for i in range(len(c)):
        print(f"Unique values in {c[i]} are :{df[c[i]].unique()}")
        print(f'The Mode of the col {c[i]} is  {df[c[i]].mode()[0]}')
        print()
        
def info_num(c):
    for i in range(len(c)):
        print(c[i])
        print(f'The Mean is : {df[c[i]].mean()}')
        print(f'The Median is : {df[c[i]].median()}')
        print(f'The STD is : {df[c[i]].std()}')
        print()
        
        
info_num(numerical)        
info_cat(categorical)
```

```python
# Task2

for i in range(len(categorical)):
    for j in range(i+1,len(categorical)):
        if categorical[i] != categorical[j]:
            print(categorical[i],"vs",categorical[j])
            plt.figure(figsize=(8,4))
            sns.countplot(y=categorical[i],hue=categorical[j],data=df)
            plt.show()
            print()
            
    
```
