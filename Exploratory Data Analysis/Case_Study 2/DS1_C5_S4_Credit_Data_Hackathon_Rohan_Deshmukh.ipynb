```python
import matplotlib.pyplot as plt            
from tabulate import tabulate               
import matplotlib as mat                    
import seaborn as sb                       
import pandas as pd                        
import numpy as np 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('DS1_C5_S4_Credit_Data_Hackathon.csv')
df
```

```python
df.info()
```

```python
round(((df.isnull().sum()*100)/df.shape[0]),2)
```

```python
# replacing Categorical value with 'not mentioned' and numerical value with'0' 
# to check if person who donot fill value are the defaulters or not
df.OCCUPATION_TYPE.fillna('Not Mentioned',inplace=True)
df.NAME_TYPE_SUITE.fillna('Not Mentioned',inplace=True)
df.AMT_GOODS_PRICE.fillna(0,inplace=True)
round(((df.isnull().sum()*100)/df.shape[0]),2)
```

```python
def col(df):
    x=df.select_dtypes(exclude='object')
    y=df.select_dtypes(include='object')
    x1=list(x.columns)
    y1=list(y.columns)
    return x1,y1


numerical,categorical=col(df)
print("Numerical")
print(numerical)
print("Categorical")
print(categorical)
```

```python
def info_cat(c):
    for i in range(len(c)):
        print(c[i])
        print(f"Unique values are :{df[c[i]].unique()}")
        print(f'The Mode is  {df[c[i]].mode()[0]}')
        print()
        
        
info_cat(categorical)
```

```python
def info_num(c):
    for i in range(len(c)):
        print(c[i])
        print(f'The Mean is : {df[c[i]].mean()}')
        print(f'The Median is : {df[c[i]].median()}')
        print(f'The Mode is : {df[c[i]].mode()[0]}')
        print(f'The STD is : {df[c[i]].std()}')
        print()
        
        
info_num(numerical)
```

```python
df1=df[df.TARGET==1]
df1
```

```python
plt.figure(figsize=(12,8))
plt.subplot(211)
ax = sb.countplot(x='GENDER',hue='TARGET',data=df1)
for label in ax.containers:
    ax.bar_label(label)

plt.subplot(212)
y=df1.groupby('GENDER')["GENDER"].value_counts()
plt.pie(y,labels=y.index,autopct='%.2f%%')
plt.show()
```
