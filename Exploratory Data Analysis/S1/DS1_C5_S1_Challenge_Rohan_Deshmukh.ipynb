```python
# Import all lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_excel('DS1_C5_S1_Income_Data_Challenge.xlsx')
df[:5]
```

```python
# Task 1
df.info()
```

```python
# Task2
def col(df):
    x=df.select_dtypes(exclude='object')
    y=df.select_dtypes(include='object')
    x1=list(x.columns)
    y1=list(y.columns)
    return x1,y1


numerical,categorical=col(df)
print(numerical)
print(categorical)
```

```python
# Task 3
def info_cat(c):
    for i in range(len(c)):
        print(f"Unique values in {c[i]} are :{df[c[i]].unique()}")
        print(f'The Mode of the col {c[i]} is  {df[c[i]].mode()[0]}')
        print()
        
info_cat(categorical)
```

```python
#task 4
for i in range(len(categorical)):
    print(categorical[i])
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sb.countplot(x=categorical[i],data=df)

    plt.subplot(122)
    per=df[categorical[i]].value_counts()
    l=df[categorical[i]].unique()
    plt.pie(per,labels=l,autopct='%.2f%%')
    plt.show()
    print()
```

```python
# Task 5
def info_num(c):
    for i in range(len(c)):
        print(c[i])
        print(f'The Mean is : {df[c[i]].mean()}')
        print(f'The Median is : {df[c[i]].median()}')
        print(f'The STD is : {df[c[i]].std()}')
        print(f'The Null count is : {df[c[i]].isnull().sum()}')
        
        print()

info_num(numerical)
```

```python
# Task 6
l=numerical
for i in range(len(l)):
    print(l[i])
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sb.histplot(df[l[i]])

    plt.subplot(122)
    sb.boxplot(df[l[i]])
    plt.show()
```
