```python
# Import all lib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv('DS1_C5_S1_EmployeeAttrition_Data_Practice.csv')
df[:5]
```

```python
# Task 1
# checking information about given data
df.info()
```

```python
# Task2
#differentiating intocategorical and numerical values
def col(df):
    x=df.select_dtypes(exclude='object')
    y=df.select_dtypes(include='object')
    x1=list(x.columns)
    y1=list(y.columns)
    return x1,y1


numerical,categorical=col(df)
print(numerical)
print(categorical)
```

```python
# Task 3
#getting mean median mode of categorical values
def info_cat(c):
    for i in range(len(c)):
        print(f"Unique values in {c[i]} are :{df[c[i]].unique()}")
        print(f'The Mode of the col {c[i]} is  {df[c[i]].mode()[0]}')
        print(f'The Null values of the col {c[i]} is  {df[c[i]].isnull().sum()}')
        print()
        
info_cat(categorical)
```

```python
# dropping  null values from categorical values
df.dropna(subset=categorical,inplace=True)
info_cat(categorical)
```

```python
#task 4
# using count plot
for i in range(len(categorical)):
    print(categorical[i])
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sb.countplot(x=categorical[i],data=df)

    plt.subplot(122)
    per=df[categorical[i]].value_counts()
    l=df[categorical[i]].unique()
    plt.pie(per,labels=l,autopct='%.2f%%')
    plt.show()
    print()
```

```python
# Task 5
#getting mean median mode of categorical values
def info_num(c):
    for i in range(len(c)):
        print(c[i])
        print(f'The Mean is : {df[c[i]].mean()}')
        print(f'The Median is : {df[c[i]].median()}')
        print(f'The STD is : {df[c[i]].std()}')
        print(f'The Null count is : {df[c[i]].isnull().sum()}')
        
        print()

info_num(numerical)
```

```python
for i in range(len(numerical)):
    df[numerical[i]]=df[numerical[i]].fillna(value=df[numerical[i]].mean())


info_num(numerical)
```

```python
# Task 6
l=['MonthlyIncome','HourlyRate','Age']
for i in range(len(l)):
    print(l[i])
    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sb.histplot(df[l[i]])

    plt.subplot(122)
    sb.boxplot(df[l[i]])
    plt.show()
```
