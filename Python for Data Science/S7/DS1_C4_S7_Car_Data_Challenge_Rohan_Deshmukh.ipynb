```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("DS1_C4_S7_Car_Data_Challenge.csv")
data
```

```python
# Task 1
# creating list of column need to be filtered by removing nullvalues
l=['Make','Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque']
# using dropna for removing values and using inplace for atual change in dataframe
data.dropna(subset=l,inplace=True)
data
```

```python
# Task 2
#Removing any occurances of "kg" in this the column
data.Kerb_Weight=data.Kerb_Weight.str.replace(r'kg','')
# Changing datatype of column
data=data.astype({'Kerb_Weight':float})
#converting kg value to pound by mulitplying it with 2.2
data.Kerb_Weight=data.Kerb_Weight*2.2

# Regular expression used for cleaning the Power column
data["Power"] = data["Power"].str.strip () # stripping white spaces if any 
data['Power'] = data['Power'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True)
# Changing datatype of column
data=data.astype({'Power':float})

data["Torque"] = data["Torque"].str.strip() # stripping white spaces if any 
data['Torque'] = data['Torque'].str.replace(r"-",'', regex=True) # Removing any occurances of "-" in this the column
data['Torque'] = data['Torque'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True)
# Changing datatype of column
data=data.astype({'Torque':float})

print(data.Kerb_Weight[:1])
print(data.Power[:1])
print(data.Torque[:1])
```

```python
# Task 3
# creating list for unique value in make column
l1=list(data.Make.unique())
# creating list for the column we need for calculation
l2=['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque']
l=['Make','Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque']
x=data[l] # creating dataframe with specific columns

for i in range(len(l1)): # looping the data because we need same result for another manufactorer in same dataframe
    df=pd.DataFrame()# creating new dataframe for output
    df[l1[i]]=l2 # craeting rows 
    df.set_index(l1[i],inplace= True) # changing index
    
    z=x[(x['Make']==l1[i])] # giving condition to Dataframe
    z.drop(['Make'],axis=1,inplace=True) # removing column since its not need 
    
    # getting mean, median, standard deviation and coefficient of variance
    df["Mean"]=z.mean()
    df["Median"]=z.median()
    df["STD_Dev"]=z.std()
    df["CV in %"]=np.round((st.variation(z)*100),2)
    print(df)
    plt.title(l1[i])
    z.Displacement.plot(kind='box', showmeans=True)# creating box plot
    plt.show()
    print()
```

```python
# Task 4
# creating list for company name
l3=['Renault','Toyota','Honda']
for i in range(len(l3)):# looping the data because we need same result for another manufactorer in same dataframe
    a=data[data["Make"]==l3[i]] # giving condition to dataframe
    print(l3[i])
    
    # getting correalatinn between milage and displacement
    co1=round((a.City_Mileage_km_litre.corr(a.Displacement)),2)
    plt.xlabel("City_Mileage_km_litre")
    plt.ylabel("Displacement")
    plt.title("City_Mileage vs Displacement")
    # Plotting scatterplot with label for correlation value
    plt.scatter(a.City_Mileage_km_litre,a.Displacement,label="Correlation="+str(co1))
    plt.legend()
    plt.show()
    
    # getting correalatinn between milage and torque
    co2=round((a.City_Mileage_km_litre.corr(a.Torque)),2)
    plt.xlabel("City_Mileage_km_litre")
    plt.ylabel("Torque")
    plt.title("City_Mileage vs Torque")
    # Plotting scatterplot with label for correlation value
    plt.scatter(a.City_Mileage_km_litre,a.Torque,label="Correlation="+str(co2))
    plt.legend()
    plt.show()
    
    # getting correalatinn between milage and power
    co3=round((a.City_Mileage_km_litre.corr(a.Power)),2)
    plt.xlabel("City_Mileage_km_litre")
    plt.ylabel("Power")
    plt.title("City_Mileage vs Power")
    # Plotting scatterplot with label for correlation value
    plt.scatter(a.City_Mileage_km_litre,a.Power,label="Correlation="+str(co3))
    plt.legend()
    plt.show()
```
