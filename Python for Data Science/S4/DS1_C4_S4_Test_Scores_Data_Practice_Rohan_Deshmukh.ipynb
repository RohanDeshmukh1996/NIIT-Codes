```python
import pandas as pd
data=pd.read_csv("DS1_C4_S4_Test_Scores_Data_Practice.csv")
data
```

```python
# Task 1
# using is to find null values and then doing sum for total count of null values in each column
print(data.isnull().sum())
print(data.test2.value_counts())
x=round(data.test1.mean(),1) # getting mean
y=data.test2.value_counts().head(1).index[0]# getting mode

# replacing null value using fillna with mean of specific columns and mode of that specific column
data.test1=data.test1.fillna(x) 
data.test2=data.test2.fillna(y)
```

```python
# running above code for verification 
print(data.isnull().sum())
data.test2.value_counts()
```

```python
# Task 2
# deleting duplicate values using drop duplicates  
data.drop_duplicates(inplace=True)
data
```

```python
# Task 3
o=["Qualifies for reduced/free ","Does not qualify"]
c=["Yes","No"]

# using replace function to replace values with new values
data.tuitionfee_waiver=data.tuitionfee_waiver.replace(to_replace=o,value=c)
data
```

```python
# Task 4
# using group by function to make groupk by school and shown average marks in test1 in descending order
data.groupby('school')['test1'].mean().sort_values(ascending=False)
```

```python
# Task5
# using pivot table to group by school with top 5 
data.pivot_table(index=['school'],values=['test2'], aggfunc="mean").sort_values(by="test2",ascending=False).head()
```

```python
# Task 6
# giving condition and storing index in object x
x=data[(data.school=="FBUMG")&(data.classroom=="5LQ")&(data.test2>=80.0)].index
# storing index of original data in object index
index=data.index

# creating Series for new column we need to add in original data with value as 0 
s=pd.Series(0,index=index)
# using xwe make changes in series 
s[x]=500
ucing concat to add new column in original data
data=pd.concat([data,s.rename("scholarship")],axis=1)
data
```

```python
# Task 7
# creating cross table for required output
pd.crosstab(data.school_setting,data.teaching_method)
```

```python
# Task 8
#using group by function on school setting and gender and use count function on gender
data.groupby(['school_setting','gender'])['gender'].count()
```

```python
# Task 9
#using group by function on school setting and gender and use mean function on test2
data.groupby(['school_setting','gender'])['test2'].mean()
```
