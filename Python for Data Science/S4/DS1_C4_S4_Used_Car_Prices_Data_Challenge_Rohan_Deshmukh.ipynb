```python
import pandas as pd
data=pd.read_csv("DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv")
data
```

```python
# Task 1
print(data.isnull().sum())
#Using dropna to remove all the rows who have null values by using inplace 
data.dropna(inplace=True)
data.isnull().sum()
```

```python
# Task 2
# giving condition and sorting the values acending and only saving top 100 in new dataframe
x=data[(data.year==2014)&(data.state=="ca")].sort_values("mp").head(100)
x
```

```python
#Task 3
# using groupby made groub of make column and used mesn on mp column
x.groupby('make')['mp'].mean()
```

```python
# Task 4
# using condition on dataframe and storing it in new data frame
y=x[x.make=="Chevrolet"]
# adding new column in new dataframe in which discount value is shown
y['discount_price']=y.mp-(y.mp*0.1)
y
```

```python
# Task 5
# using pivot table  fro interior color and men function on mp
x.pivot_table("mp","interior",aggfunc=['mean'])
```

```python
# Task 6
# giving condiition and storing it in new data frame
y=data[(data.year>=2010)&(data.year<=2015)]
# using groupby on new dataframe  for make, color and count on color for values
y.groupby(['make','color'])['color'].count()
```

```python
# Task 7
# giving condiition and storing it in new data frame
y=data[(data.make=="BMW")&((data.body=="Sedan")|(data.body=="sedan"))&(data.year>=2010)&(data.year<=2015)
     &(data.condition>=4.5)]
# using pivot table on new dataframe  for year, model and max on mp for values
y.pivot_table(index=['year','model'],values=['mp'], aggfunc='max')
```

```python
# Task 8
# giving condiition and displaying it
data[(data.body=="SUV")&(data.year>=2010)&(data.year<=2015)&(data.odometer<=30000)].sort_values(ascending=False,by="odometer").head(100)
```
