```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
```

```python
data=pd.read_excel('DS1_C4_S5_Car_Data_Challenge.xlsx')
data
```

```python
# Task 1
# using groupby for make and mean function on city milage as value and saving in object
y=data.groupby("Make")["City_Mileage_km_litre"].mean().sort_values(ascending=False).head(10)
# plotting Horizontal bar graph
plt.barh(y.index,y)
plt.title("horizontal bar chart")
plt.show()
```

```python
# Task 2
# giving condition to data frame and storing it in new dataframe
z=data[data.Fuel_Type!="Electric"]
# using groupby for make and mean function on city milage as value and saving in object
y=z.groupby("Make")["City_Mileage_km_litre"].mean().sort_values(ascending=False).head(10)
# plotting Horizontal bar graph
plt.barh(y.index,y)
plt.show()
```

```python
# Task 3
# using groupby for body type and count function on body type as value and saving in object
y=data.groupby(['Body_Type'])['Body_Type'].count().sort_values(ascending=False).head()
# plotting Horizontal bar graph
plt.barh(y.index,y)
plt.show()
```

```python
# Task 4
# giving condition to data frame and storing it in new dataframe
x=data[(data.Make=='Hyundai')|(data.Make=='Mahindra')|(data.Make=='Renault')|(data.Make=='Skoda')]
# using pivot table getting mean of car makers with respect to city milage and highway milage 
a=x.pivot_table(index=['Make'],values=['City_Mileage_km_litre'], aggfunc=['mean'])
print(a)
c=x.pivot_table(index=['Make'],values=['Highway_Mileage_km_litre'], aggfunc=['mean'])
print(c)
# using pivot table getting meadian of car makers with respect to city milage and highway milage 
b=x.pivot_table(index=['Make'],values=['City_Mileage_km_litre'], aggfunc=['median'])
print(b)
d=x.pivot_table(index=['Make'],values=['Highway_Mileage_km_litre'], aggfunc=['median'])
print(d)
```

```python
# Task 5
# using pivot table getting mean and median of car makers with respect to city milage
a=x.pivot_table(index=['Make'],values=['City_Mileage_km_litre'], aggfunc=['mean','median'])
# ploting stacked bar graph 
a.plot(kind='bar', stacked=True)
plt.show()
```

```python
# Task 6
# storing mean values in object
x=round(data.City_Mileage_km_litre.mean(),2)
y=round(data.Highway_Mileage_km_litre.mean(),2)
#using fillna to fill null values with mean value
data.City_Mileage_km_litre=data.City_Mileage_km_litre.fillna(x) 
data.Highway_Mileage_km_litre=data.Highway_Mileage_km_litre.fillna(y)
# using pivot table getting mean of car makers with respect to city milage and highway milage 
a=data.pivot_table(index=['Make'],values=['City_Mileage_km_litre','Highway_Mileage_km_litre'], aggfunc=['mean'])

#ploting horizontal bar graph
plt.barh(a.index,y)
plt.show()
```

```python
# Task 7
# using sort function sort valus in decending order 
a=data.sort_values(by='City_Mileage_km_litre',ascending=False).head(10)
# plotting line graph
plt.plot(a.Displacement,a.City_Mileage_km_litre,marker='o',ls='--',color="red")
plt.title("Line Graph")
```
