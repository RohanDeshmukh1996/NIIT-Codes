## Task 10

```python
ext=df.sort_values('revenue', ascending=False).head(500)
# creating new dataframe for output 
ct_df=pd.DataFrame()
ct_df['Central_Tend']=['Mean','Median','Mode']# adding values in data frame
ct_df.set_index("Central_Tend",inplace= True)# changing the index

l=['budget','revenue','runtime']
for i in range(len(l)):
    a=[]
    # getting mean median mode and storing in list
    a.append(ext[l[i]].mean())
    a.append(ext[l[i]].median())
    a.append(ext[l[i]].mode()[0])
    ct_df[l[i]]=a #adding list into data frame 

ct_df
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
sb.boxplot(y=ext['budget'], showmeans= True)
plt.subplot(1,3,2)
sb.boxplot(y=ext['revenue'], showmeans= True)
plt.subplot(1,3,3)
sb.boxplot(y=ext['runtime'], showmeans= True)
plt.show()
```
