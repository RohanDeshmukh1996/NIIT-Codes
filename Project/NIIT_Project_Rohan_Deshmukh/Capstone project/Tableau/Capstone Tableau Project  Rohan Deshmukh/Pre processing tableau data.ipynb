## Null value finding and removal

```python
df.isnull().sum()
```

```python
df.Credit_Limit.fillna(df.Credit_Limit.median(),inplace=True)
df.Card_Category.fillna(df.Card_Category.mode()[0],inplace=True)
df.isnull().sum()
```

## Outlier analysis of numerical values

```python
numeric=['Customer_Age', 'Dependent_count', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Trans_Amt', 'Total_Trans_Ct',
       'Avg_Utilization_Ratio']
data=df[numeric]

iqr_df=pd.DataFrame()
iqr_df["Columns"]=numeric
iqr_df.set_index("Columns",inplace= True)
# getting mean median range and storing in data frame
iqr_df["Min"]=data.min()
iqr_df["Max"]=data.max()
iqr_df["Range"]=data.max()-data.min()
# using quantile to get Q1 and Q3
iqr_df["Q1"]=data.quantile(0.25)
iqr_df["Q3"]=data.quantile(0.75)
iqr_df["IQR"]=iqr_df["Q3"]-iqr_df["Q1"]
iqr_df["LF"]=iqr_df["Q1"]-(1.5*iqr_df["IQR"])
iqr_df["UF"]=iqr_df["Q3"]+(1.5*iqr_df["IQR"])

count=[]
def outlier_count(a,LF,UF):
    x=0
    y=0
    for i in data[a]: 
        if i<LF:            
            x=x+1   
        elif i>UF:         
            y=y+1
    count.append((x+y))
i=0       
while i<12:
    outlier_count(numeric[i],iqr_df.LF[i],iqr_df.UF[i])
    i+=1
    
iqr_df["Count_Outliers"]=count
iqr_df
```
